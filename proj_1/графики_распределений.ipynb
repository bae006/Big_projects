```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('выборка_после_кластеризации.csv')
df
```

```python
df.columns
```

```python
for i in ['EXT_SOURCE_3', 'EXT_SOURCE_2', 'EXT_SOURCE_1', 'month_employed',\
          'higher_EDUCATION_TYPE', 'CODE_GENDER', 'AMT_CREDIT', 'REGION_RATING_CLIENT_W_CITY',\
          'age', 'OWN_CAR_AGE', 'communication_count']:

    plt.figure()
    df[i].hist(color='blue', bins = 30, label="hist")
    plt.title(f'Общее распределение')
    #plt.xlim(0, 10)
    plt.show()

    
    df1 = df.loc[df["TARGET"] == 1]
    df0 = df.loc[df["TARGET"] == 0]


    # Строим график для target = 1 и target = 0
    print("-------------------------------", i)
    plt.figure()
    df1[i].hist(color='red', label='Target = 1', alpha = 1, bins = 30, density=True)
    df0[i].hist( color='blue', label='Target = 0', alpha = 0.5, bins = 30, density=True)
    plt.title(f'Распределение с TARGET = 1 и TARGEN = 0')
    #plt.xlim(0, 10)
    plt.legend()
    plt.show()
```
